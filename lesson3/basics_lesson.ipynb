# Loops
- These let us run code however many times we need
- This is extremely useful when solving larger problems or creating continuously running programs (Ex. video games)

# For loop
- A loop that runs a certain amount of times

```python
loop_times = 10

for counter in range(loop_times):
    print("This is loop number " + str(counter))

print('Done.')
```

# While loop
- A loop that runs so long as the condition is fulfilled

```python
counter = 0

while counter < 10:
    print("This is loop number " + str(counter))
    counter += 1

print('Done.')
```

# Infinite loops
- These will run continuously unless there is a built in stopper or it is forcefully stopped. This kind of loop will be extremely useful later on

```python
import time

while True:
    print("This loop will run forever!")
    
    # We put a small delay so that it doesn't use up all of our computer's resources, since this loop will run forever.
    time.sleep(1)
```
