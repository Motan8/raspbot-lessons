# Lists
- In this lesson we will learn how to use lists.
- They are a data type that allows us to store data inside of. They are extremely useful for handling info on a large scale.
- They are also useful for smaller programs. They can improve our code efficiency.

```python
colors = ["red", "green", "blue", "yellow", "purple"]

for color in colors:
    print(color)
```

What if we wanted to number the colors?

```python
colors = ["red", "green", "blue", "yellow", "purple"]

for counter in range(len(colors)):
    print(f"Color {counter + 1}: {colors[counter]}")
```

What if we wanted to list them in reverse order?

```python
colors = ["red", "green", "blue", "yellow", "purple"]

for counter in range(len(colors) - 1, -1, -1):
    print(f"Color {counter + 1}: {colors[counter]}")
```

Other useful functions we can use on lists


```python
favourite_numbers = [7, 3, 42, 12, 5]

print(favourite_numbers[1]) # This is how you access a specific value in the list (index 1 in this case)

print('Adding 9')
favourite_numbers.append(9) # Adds a new value to the end of the list
print(favourite_numbers)

print('Adding 15 at index 2')
favourite_numbers.insert(2, 15) # Inserts a value at a specific index (2 in this case)
print(favourite_numbers)

print('Removing 3')
favourite_numbers.remove(3) # Removes a specific value from the list
print(favourite_numbers)

print('Popping list')
favourite_numbers.pop() # Removes the last value from the list and returns it (not used here)
print(favourite_numbers)

print('Popping value at index 1')
favourite_numbers.pop(1) # Removes the value at index 1 and returns it (not used here)
print(favourite_numbers)

print('Sorting list')
favourite_numbers.sort() # Sorts the list in ascending order
print(favourite_numbers)

print('Clearing list')
favourite_numbers.clear() # Removes all values from the list
print(favourite_numbers)
```
