# If & else
Conditional statements are what allow us to create more sophisticated programs.

# Basic if: 
- Code chain that only runs if the condition is satisfied.

```python
height_in_cm = 150

# check if this person can attend the roller coaster
if height_in_cm > 160:
    print("You can attend the roller coaster!")

print('Done.')
```

But what about when the person isn't tall enough?

```python
height_in_cm = 150

# check if this person can attend the roller coaster
if height_in_cm < 160:
    print("You cannot attend the roller coaster.")
else:
    print("You can attend the roller coaster!")
    
print('Done.')
```

What if the person is too tall?

```python
height_in_cm = 170

# check if this person can attend the roller coaster
if height_in_cm < 160:
    print("You cannot attend the roller coaster.")
elif height_in_cm >= 180:
    print("You cannot attend the roller coaster.")
else:
    print("You can attend the roller coaster!")

print('Done.')
```

We can simply this even further

```python
height_in_cm = 150

# check if this person can attend the roller coaster
if height_in_cm < 160 or height_in_cm > 180:
    print("You cannot attend the roller coaster.")
else:
    print("You can attend the roller coaster!")

print('Done.')
```

We can combine several statements into one line using "and" which will require both to be fulfilled, but we can also use "or" which means only one of the conditions need to be fullfiled.

If we need even more complicated conditions, we can surround them in brackets () to prioritize them, kind of like math and BEDMAS.

```python
height_in_cm = 150
safe_seat = True

# check if this person can attend the roller coaster
if (height_in_cm < 160 or height_in_cm > 180) and not safe_seat:
    print("You cannot attend the roller coaster.")
else:
    print("You can attend the roller coaster!")

print('Done.')
```

```python
The different comparison operators we can use are:

< Smaller than
> Greater than
== Equal to
!= Does not equal to
<= Smaller than or equal to
>= Greater than or equal to
```
