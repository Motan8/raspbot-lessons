# Project 2
The students need to create a program that can assist a driver in maintaining a straight path on the road.

Their program needs to:
- Use middle two infrared sensors.
- Warn the driver if the car is steering too far left or right.
  - If it is going too far right, make it buzz loudly for a second.
  - If it is going too far left, make it buzz quietly for a second.
  - If it is going straight, do nothing.
