# Controlling the car
We will get the car to create noise

There will be a lot of times where we will need to use libraries, which are pre-written code that we can use to do things. We import them at the top of our code so we can use them later on.

```python
import RPi.GPIO as GPIO
```

GPIO is what lets us control the many devices on the car. This is how we want to set it up.

```python
GPIO.setwarnings(False)
# This lets us choose pins by their physical location on the board.
GPIO.setmode(GPIO.BOARD)
# Set pin 32 to be an output pin, which lets us send signals to it.
GPIO.setup(32, GPIO.OUT) 

# PWM is a way for us to send signals. We are sending a frequency of 440 Hz. Change this to get different notes!
buzzer = GPIO.PWM(32, 440)
```

```python
# This turns on the buzzer at 29% volume.
buzzer.start(30)
buzzer.stop()
print('Done')
```

Why did the buzzer not make noise?
- Code runs very quickly on new technology. We have to force it to wait between turning the buzzer on and off!
- We will do this by importing the time library.

```python
import time
```

```python
print('Waiting 1 second')
time.sleep(1)
print('Done')
```

Adding it to our program:

```python
# This turns on the buzzer at 29% volume.
buzzer.start(30)

time.sleep(1) # <--

buzzer.stop()
# This turns off the buzzer and resets the pin so it can be used for other things later.
GPIO.cleanup()
print('Done')
```
