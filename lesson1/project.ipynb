Lets make our first project!

We will get the buzzer to share a message through morse code

```python
import RPi.GPIO as GPIO
import time
```

```python
GPIO.setwarnings(False)
GPIO.setmode(GPIO.BOARD)
GPIO.setup(32, GPIO.OUT) 

buzzer = GPIO.PWM(32, 440)
```

This program will communicate the message "Hi"
- In morse code, H is 4 short beeps, and I is 2 short beeps

```python
time_unit = 0.1

# Letter H
buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

# Wait between letters
time.sleep(time_unit * 3)

# Letter I
buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

buzzer.start(30)
time.sleep(time_unit)
buzzer.stop()
time.sleep(time_unit)

buzzer.stop()
# GPIO.cleanup()
print('Done')
```
