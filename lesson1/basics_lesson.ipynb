# Variables
Variables are really useful because they let us store information and use it later. We can also change the value of a variable by assigning it a new value.

Note: In programming, the equal sign does not mean equal. It actually means "assign"

```python
# When writing in python, we name variables using snake_case, which means we use lowercase letters and separate words with underscores. This is just a convention but it makes our code easier to read.
favourite_number = 8
my_name = "Mo"

# We can pass variables into functions, and it will fill in the variable with its value.
print(f"My name is {my_name} and my favourite number is {favourite_number}")
```

```python
# We can also do math with variables.
simple_math = 1 + 1

# We can use the print function to show the value of a variable. We don't always have to put text in there with it.
print(simple_math)
```

```python
date_of_birth = 2008
current_year = 2026

age = current_year - date_of_birth
print(f"I am {age} years old.\n") # Putting "\n" will insert a new line, like pressing enter.
```

```python
distance_to_school = 3.5 # km
walk_speed = 5 # km/h
time_to_walk = distance_to_school / walk_speed
print(f"It takes me {time_to_walk} hours to walk to school.\n")

# We can also do math inside of the print function, without needing to create a variable for it.
print(f"It takes me {distance_to_school / walk_speed} hours to walk to school.\n")
```
