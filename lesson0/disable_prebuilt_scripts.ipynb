# Lets make it so the Raspbot prioritizes other WiFi networks rather than its hotspot
(Run this only ONCE after turning on the Raspbot for the first time)
- This makes it so it would look for your WiFi before it connects to its hotspot
- Its more convenient to have it this way since it'll make things more automatic

## What to do after running it:
- Using the phone app, connect the Raspbot to the WiFi

```python
import os

os.system('sudo raspi-config nonint do_wifi_country CA')
os.system('sudo nmcli connection modify "Raspbot" connection.autoconnect-priority -10')
print('Done.')
```

# Lets set the Raspbot up to run our own code
- The Raspbot is occupied since it is waiting from a signal from the app.
- We will tell it to stop waiting so we can run our own programs.
- Do this by running the cell below. We will have to repeat this every time we turn the Raspbot on.

```python
import os

os.system('sudo pkill -9 -f raspbot.pyc')
print('Done.')
```
